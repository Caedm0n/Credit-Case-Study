### Data Cleaning and Handling

```python
#Calculating missing values percentage in previous application dataset

missingValues = 100*previous_application_data.isnull().sum()/previous_application_data.shape[0]
missingValues = missingValues.to_frame()
missingValues.reset_index(level=0,inplace=True)
missingValues.columns = ['Column','Null_Percent']
```

```python
missingValues[missingValues.Null_Percent>0].sort_values(by='Null_Percent',ascending=False)
```



*   We can drop columns **RATE_INTEREST_PRIVILEGED**, **RATE_INTEREST_PRIMARY** as they have almost all data missing (99%)
*   We can drop rows containing null values for **PRODUCT_COMBINATION**, **AMT_CREDIT** as the percentage of null values is less than 10% 



```python
#Dropping entire column where number of null are greater.
previous_application_data.drop(['RATE_INTEREST_PRIMARY','RATE_INTEREST_PRIVILEGED'], axis=1, inplace=True)

#Dropping rows where number of nulls is less than 10%
previous_application_data.dropna(subset=['PRODUCT_COMBINATION','AMT_CREDIT'], inplace=True)
```

```python
#Identifying numeric and categorical columns
numeric_columns = []
categorical_columns = []
for i in previous_application_data.columns:
  if previous_application_data[i].dtype in ['float64','int64']:
    numeric_columns.append(i)
  else:
    categorical_columns.append(i)
```

```python
numeric_columns
```

```python
categorical_columns
```

```python
#Imputing missing values for numeric and catergorical data
for i in missingValues.values:
  if i[0] in numeric_columns and i[1]>0.0:
    previous_application_data[i[0]].fillna(previous_application_data[i[0]].median(),inplace=True)
  elif i[0] in categorical_columns:
    previous_application_data[i[0]].fillna(previous_application_data[i[0]].mode()[0], inplace=True)
```

**Finding Category wise Defaulters in merged dataset**


* NAME_CASH_LOAN_PURPOSE
* NAME_CONTRACT_STATUS
* NAME_PAYMENT_TYPE
* NAME_TYPE_SUITE
* NAME_CLIENT_TYPE
* NAME_GOODS_CATEGORY
* NAME_PORTFOLIO
* NAME_PRODUCT_TYPE
* CHANNEL_TYPE
* NAME_SELLER_INDUSTRY



```python
def category_defaulter_percentage(df, col):
    new_df = pd.DataFrame(columns=['Column', 'Percentage of Defaulter'])
    for i in df[col].unique():
        defaults = df[(df[col] == i) & (df.TARGET == 1)].shape[0]
        total = df[df[col] == i].shape[0]
        new_df = new_df.append({'Column' : i , 'Percentage of Defaulter' : (defaults*100/total)}, ignore_index=True)

    new_df.set_index('Column',inplace = True)   
    return new_df.sort_values(by='Percentage of Defaulter', ascending=False).plot.bar(figsize=(10,5))
```

```python
category_defaulter_percentage(merged_data,'NAME_CASH_LOAN_PURPOSE')
plt.show()
```

```python
category_defaulter_percentage(merged_data,'NAME_CONTRACT_STATUS')
plt.show()
```
